```python
import matplotlib.pyplot as plt
import numpy as np
import math

#returns value of a[0]x(n-1)+a[1]x(n-2)+...+a[n-1]
def horner(a, n, x):
    #Initialize result p
    p = a[0]
    
    #Evaluate value with Horners algorithm
    for i in range(1,n):
        p = p*x+a[i]
        #print(p)
    return p

#p2(x)=x^9 − 18x^8 + 144x^7 − 672x^6 + 2016x^5 − 4032x^4 +... 
#5376x^3 − 4608x^2 + 2304x − 512
a = [1, -18, 144, -672, 2016, -4032, 5376, -4608, 2304, -512]
x = 9
n = len(a)
#8000
bob = np.linspace(1.92,2.08, num=8000)
k = len(bob)
tmp = np.empty(k, dtype=object)

for i in range(0,k):
    tmp[i] = horner(a, n, bob[i])
y = tmp
print("Values 1-200:")
for i in range(0,201):
    print(y[i])
print("Values 200-500:")
for i in range(200,501):
    print(y[i])

plt.plot(bob, y)
plt.title('P2(x)')
plt.show()
```
